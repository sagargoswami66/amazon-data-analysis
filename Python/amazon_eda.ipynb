# FInd null values if exist

```python
print(df.isnull().sum())
```

# Remove the null values

```python
if 'Rating Count' in df.columns:
    df['Rating Count'].fillna(0, inplace=True)
```

# Fix the type of 'discount_percentage' column

```python
df['discount_percentage'] = (
    df['discount_percentage']
    .str.replace('%', '', regex=False)
    .astype(float)
)

```

# Fix the type

```python
price_cols = ['discounted_price', 'actual_price']

for col in price_cols:
    df[col] = (
        df[col]
        .str.replace('â‚¹', '', regex=False)
        .str.replace('Rs.', '', regex=False)
        .str.replace('Rs', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip()
    )
    df[col] = pd.to_numeric(df[col], errors='coerce')
```

# Fix the type and fill the null values with 0

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df['rating'].fillna(0, inplace=True)
```

# Fix the type and fill the null values with 0

```python
df['rating_count'] = df['rating_count'].astype(str)
df['rating_count'] = df['rating_count'].str.replace(',', '', regex=False)
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
df['rating_count'].fillna(0, inplace=True)
df['rating_count'] = df['rating_count'].astype(int)
```

# Dropping unnecessary colums

```python
# List of columns to drop
columns_to_drop = ['about_product', 'review_title', 'review_content', 'img_link', 'product_link']

# Drop columns
df = df.drop(columns=columns_to_drop, errors='ignore')
```

# Save the cleaned file

```python
df.to_csv("amazon_cleaned.csv", index=False)
```

# Univariate Analysis

```python
# Rating distribution
sns.histplot(df['rating'], bins=5)
plt.title('Distribution of Rating')
plt.show()

# Rating count distribution
sns.histplot(df['rating_count'], bins=30)
plt.title('Distribution of Rating Count')
plt.show()

# Discount percentage
sns.histplot(df['discount_percentage'], bins=20)
plt.title('Distribution of Discount Percentage')
plt.show()
```

# Bivariate Analysis

```python
# Discount vs rating
sns.scatterplot(x='discount_percentage', y='rating', data=df)
plt.title('Discount vs Rating')
plt.show()

# Rating vs rating_count
sns.scatterplot(x='rating_count', y='rating', data=df)
plt.title('Rating vs Rating Count')
plt.show()
```

# Multivariate Analsis

```python
sns.heatmap(df[['rating', 'rating_count', 'discount_percentage']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
